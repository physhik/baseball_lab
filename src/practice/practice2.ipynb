The following examples are trying to see correlations of the new stats (nFIP, nxFIP and so on) and ERA or RA compared to traditional FIP, xFIP, SIERA. The understanding to it will be discussed in my personal blog.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from scipy.stats import pearsonr as coef

dps = {}
for year in range(2015, 2019):
  print("===================================================")
  print("year:" , year)
  n_year = year +1
  train=new_metric(year)
  test = new_metric(n_year)
  ids = []
  for id in train.index:
    if id in test.index:
      ids.append(id)
      
  train = train.loc[ids]
  test = test.loc[ids]
  
  train = train.sort_values(by='mlb_id')
  test = test.sort_values(by='mlb_id')
  
  tr_ip = train.IP.values
  te_ip = train.IP.values
  
  ip = []
  for i in range(len(train)):
    ip.append(min(tr_ip[i], te_ip[i]))
    
    
  trains = []
  
  trainlist = [ 'ERA', 'FIP','nFIP', 'nFIPwo', 'xnFIP', 'xnFIPwo', 'xFIP','SIERA']
  
  trains.append(train.ERA.values)
  trains.append(train.FIP.values)
  trains.append(train.nFIP.values)
  trains.append(train.nFIPwo.values)
  trains.append(train.xnFIP.values)
  trains.append(train.xnFIPwo.values)
  trains.append(train.xFIP.values)
  trains.append(train.SIERA.values)
  
  
  test = test.ERA.values
  
  analysis = {}
  for i in range(len(trains)):
    analysis[i]=[]
  
  #mse0 = sum((3.15-test)**2)/len(test)
  #print("mse0:", mse0)
  for i in range(len(trains)):
    print("mse{}:".format(trainlist[i]),sum((trains[i]-test)**2)/len(test))
    analysis[i].append(sum((trains[i]-test)**2)/len(test))
    val = sum((trains[i]-test)*(trains[i]-test)*ip*ip)/(sum(trains[i]*ip)*len(test))
    print("inning-weighted-mse-{}".format(trainlist[i]), val)
    analysis[i].append(val)




  regr = linear_model.LinearRegression()
  for i in range(len(trains)):
    print("pearson R of {}: ".format(trainlist[i]), coef(trains[i],test))
    analysis[i].append(coef(trains[i],test))

    X = trains[i].reshape(len(trains[i]),1)
    y= test.reshape(len(test),1)
    regr.fit(X, y)

    print("coef:",regr.coef_, "intercept:", regr.intercept_)

    pred = regr.predict(X)
    plt.scatter(X, y)
    plt.plot(X, pred)
    plt.ylim(0, 10)
    plt.xlim(0, 10)
    plt.show()
    print("mse:",mean_squared_error(X, pred))
    #analysis[i].append(mean_squared_error(X, pred))
    print("r2:",r2_score(X, pred))
    #analysis[i].append(r2_score(X, pred))
    print("std :",np.std(X))
    analysis[i].append(np.std(X))
    print("==============================")
  
  dp = pd.DataFrame(analysis)
  dps[year] = dp 
```
